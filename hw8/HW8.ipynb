#API

```python
#API
import requests
import pandas as pd
```

```python
base_url="https://www.anapioficeandfire.com/api"
```

```python
#Получить информацию обо всех книгах и сохранить в pd.DataFrame
book_list= requests.get(
    f'{base_url}/books').json()
len(book_list)
```

```python
df_book = pd.DataFrame(book_list)
df_book.head()
```

```python
#Получить информацию обо всех домах Вестероса (Houses) и сохранить в pd.DataFrame
houses_list= requests.get(
    f'{base_url}/houses').json()
len(houses_list)
```

```python
df_house = pd.DataFrame(houses_list)
df_house.head()
```

```python
#Получить информацию обо всех домах Вестероса, у которых есть девиз, и сохранить в pd.DataFrame
data=[]
params = {
    "hasWords" : "true"
}
```

```python
haswords=df_house[df_house['words'].str.len()>0]['name']
haswords
```

```python
haswords_list = requests.get(
    f'{base_url}/houses',
    params=params).json()
haswords_list
```

```python
df_haswords = pd.DataFrame(haswords_list)
df_haswords.head()
```
