```python

import requests #библиотека для отправки HTTP-запросов
from bs4 import BeautifulSoup #библиотека для парсинга HTML и XML документов
import pandas as pd
from tqdm import tqdm
```

```python
base_url = 'https://lifehacker.ru/topics/technology/'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}
parsed_urls=[]
response = requests.get('https://lifehacker.ru/topics/technology/?page=1', headers=headers)
soup = BeautifulSoup(response.text, 'html')
```

```python
raw_items = soup.find_all('a', class_='lh-article-card-controls__comments')  # ищем все a-элементы с классом tm-article-comments-counter-link__link
links = [item.get('href').replace('/#comments', '') for item in raw_items] # получаем у ссылок только href-атрибут, убираем подстроку '/comments', Заменяя на пустую строку
links
```

```python
for page_num in range(1, 11):
    url = f'{base_url}/?page={page_num}' # подставляем нужный номер страницы для пагинации, дальше парсим каждую страницу аналогично коду выше
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'lxml')
    raw_items = soup.find_all('a', class_='lh-article-card-controls__comments')
    parsed_urls += [item.get('href').replace('/#comments', '') for item in raw_items]
```

```python
result = []

for url in tqdm(parsed_urls):
    article = {}
    article_url = f'{base_url}{url}' # подставляем часть ссылки, ведущую материал
    response = requests.get(article_url)
    soup = BeautifulSoup(response.text, 'lxml')


    title_element = soup.find('h1', class_='article-card__title')
    if title_element:
      article['title'] = title_element.get_text(strip=True)
    else:
      article['title'] = None

    content_element = soup.find('div', class_='single-article__post-content single-article__content-container')
    if content_element:
       article['text'] = content_element.get_text(strip=True)
    else:
      article['text'] = None



    result.append(article)
```
