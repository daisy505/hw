```python
#Выведите самое популярное мужское и самое популярное женское имя на корабле
def first_name_only(full_name):
  name_part = full_name.split(",")[1].strip()
  first_name = name_part.split()[1]
  return first_name
```

```python
df['FirstNameOnly']= df['Name'].apply(first_name_only)
df
```

```python
male_names = df[df['Sex'] == 'male']['FirstNameOnly']
most_common_male_name = male_names.value_counts()

# Выделим самое популярное имя среди женщин
female_names = df[df['Sex'] == 'female']['FirstNameOnly']
most_common_female_name = female_names.value_counts()

# Выведем результаты
print(f"Самое популярное мужское имя: {most_common_male_name}")
print(f"Самое популярное женское имя: {most_common_female_name}")
```
