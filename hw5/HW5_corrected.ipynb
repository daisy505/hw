# Оптимизация типов pandas

```python
df2 = pd.read_csv("/content/Housing.csv")
```

```python
df2.head()
```

```python
#Для каждого столбца определите оптимальный с точки зрения потребления памяти тип данных
df2.info()
```

```python
df2.isna().sum() #no NAs
```

```python
df2.describe()
```

```python
#опираясь на минимальные и максимальные значения параметра, выберем подходящие типы данных
df2['price'] = df2['price'].astype(np.int32)
df2['area'] = df2['area'].astype(np.int16)
df2['bedrooms'] = df2['bedrooms'].astype(np.int8)
df2['bathrooms'] = df2['bathrooms'].astype(np.int8)
df2['stories'] = df2['stories'].astype(np.int8)
df2['parking'] = df2['parking'].astype(np.int8)
df2['mainroad'].info(memory_usage="deep") #от строковых данных переходим к категориальным значениям
df2['guestroom'].info(memory_usage="deep")
df2['basement'].info(memory_usage="deep")
df2['hotwaterheating'].info(memory_usage="deep")
df2['airconditioning'].info(memory_usage="deep")
df2['prefarea'].info(memory_usage="deep")
df2['furnishingstatus'].info(memory_usage="deep")
df2.info()
```

```python
#меняем int64 в int8 и от строковых данных переходим к категориальным значениям
# оптимизируем память, занимаемую датасетом, примерно в 1,5 раза
```
